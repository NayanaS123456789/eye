```python
import os
import torch
import torchvision.transforms as transforms
from torch.utils.data import Dataset
from PIL import Image
import xml.etree.ElementTree as ET

class VOCDataset(Dataset):
    def __init__(self, image_dir, annotation_dir, transform=None, label_map=None):
        self.image_dir = image_dir
        self.annotation_dir = annotation_dir
        self.image_filenames = [
            f for f in os.listdir(image_dir) if f.endswith(".jpg")
            and os.path.exists(os.path.join(annotation_dir, f.replace(".jpg", ".xml")))
        ]  # ✅ Only keep images that have annotations

        self.transform = transform

        # ✅ Use predefined label mapping to ensure consistency
        self.label_map = label_map if label_map else {
        "100IndianRupees": 1,	"10IndianRupees":2,"10RupeesIndianCoin":3,"2000 Rupees":4,"200IndianRupees":5,"20IndianRupees":6,"20RupeesIndianCoin":7,"500IndianRupees":8,"50IndianRupees":9,	"5IndianRupees":10,	"5RupeesIndianCoin":11,	"undefined":12
        }

    def parse_voc_xml(self, xml_path):
        """Parses a VOC XML file and extracts bounding boxes & labels."""
        tree = ET.parse(xml_path)
        root = tree.getroot()

        boxes = []
        labels = []

        for obj in root.findall("object"):
            label = obj.find("name").text
            if label not in self.label_map:  # ✅ Ignore unknown labels
                continue

            bbox = obj.find("bndbox")
            xmin = float(bbox.find("xmin").text)
            ymin = float(bbox.find("ymin").text)
            xmax = float(bbox.find("xmax").text)
            ymax = float(bbox.find("ymax").text)

            boxes.append([xmin, ymin, xmax, ymax])
            labels.append(self.label_map[label])  # ✅ Use predefined label index

        return torch.tensor(boxes, dtype=torch.float32), torch.tensor(labels, dtype=torch.int64)

    def __getitem__(self, idx):
        img_filename = self.image_filenames[idx]
        img_path = os.path.join(self.image_dir, img_filename)
        xml_path = os.path.join(self.annotation_dir, img_filename.replace(".jpg", ".xml"))

        # ✅ Load image safely
        image = Image.open(img_path).convert("RGB")

        # ✅ Parse annotation safely
        boxes, labels = self.parse_voc_xml(xml_path)

        if len(boxes) == 0:  # ✅ Handle missing bounding boxes
            target = {"boxes": torch.empty((0, 4), dtype=torch.float32), "labels": torch.empty((0,), dtype=torch.int64)}
        else:
            target = {"boxes": boxes, "labels": labels}

        if self.transform:
            image = self.transform(image)

        return image, target

    def __len__(self):
        return len(self.image_filenames)
```

```python
import os
import torch
import torchvision.transforms as transforms
import xml.etree.ElementTree as ET
from PIL import Image
from torch.utils.data import Dataset, DataLoader

# Label mapping (modify according to your dataset)
LABEL_MAP = {
    "100IndianRupees": 1, "10IndianRupees": 2, "10RupeesIndianCoin": 3, "2000 Rupees": 4,
    "200IndianRupees": 5, "20IndianRupees": 6, "20RupeesIndianCoin": 7, "500IndianRupees": 8,
    "50IndianRupees": 9, "5IndianRupees": 10, "5RupeesIndianCoin": 11, "undefined": 12
}

class VOCDataset(Dataset):
    def __init__(self, image_dir, annotation_dir, transform=None):
        self.image_dir = image_dir
        self.annotation_dir = annotation_dir
        self.image_files = [f for f in os.listdir(image_dir) if f.endswith(".jpg")]
        self.transform = transform

    def parse_voc_xml(self, xml_file):
        tree = ET.parse(xml_file)
        root = tree.getroot()

        boxes = []
        labels = []

        for obj in root.findall("object"):
            label = obj.find("name").text
            if label not in LABEL_MAP:
                continue  # Skip unknown labels

            bbox = obj.find("bndbox")
            xmin = int(bbox.find("xmin").text)
            ymin = int(bbox.find("ymin").text)
            xmax = int(bbox.find("xmax").text)
            ymax = int(bbox.find("ymax").text)

            # Ensure valid bounding boxes
            if xmin >= xmax or ymin >= ymax:
                print(f"⚠️ Invalid bounding box in {xml_file}: {xmin, ymin, xmax, ymax}")
                continue

            boxes.append([xmin, ymin, xmax, ymax])
            labels.append(LABEL_MAP[label])

        if len(boxes) == 0:
            print(f"⚠️ No valid bounding boxes found in: {xml_file}")

        return torch.tensor(boxes, dtype=torch.float32), torch.tensor(labels, dtype=torch.int64)

    def __getitem__(self, idx):
        img_filename = self.image_files[idx]
        img_path = os.path.join(self.image_dir, img_filename)
        xml_filename = os.path.splitext(img_filename)[0] + ".xml"
        xml_path = os.path.join(self.annotation_dir, xml_filename)

        image = Image.open(img_path).convert("RGB")
        boxes, labels = self.parse_voc_xml(xml_path)

        if boxes.shape[0] == 0:
            print(f"⚠️ Skipping {img_filename} due to missing annotations.")
            return None  # Skip empty annotations

        target = {"boxes": boxes, "labels": labels}

        if self.transform:
            image = self.transform(image)

        return image, target

    def __len__(self):
        return len(self.image_files)

# Define transformations
transform = transforms.Compose([
    transforms.Resize((320, 320)),  # Resize for SSD
    transforms.ToTensor()
])

# Create datasets
train_dataset = VOCDataset("dataset/train/images", "dataset/train/annotations", transform=transform)
valid_dataset = VOCDataset("dataset/valid/images", "dataset/valid/annotations", transform=transform)

# DataLoader with collate function
def collate_fn(batch):
    batch = [b for b in batch if b is not None]
    return tuple(zip(*batch)) if batch else None

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=collate_fn)
valid_loader = DataLoader(valid_dataset, batch_size=8, shuffle=False, collate_fn=collate_fn)

print("✅ Data loaders created successfully!")
```

```python
import torch.optim as optim
import torch
from torchvision.ops import box_iou
from torch.utils.data import DataLoader

# Define optimizer and loss function with weight decay to reduce overfitting
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.5)

# Training loop
num_epochs = 15

# Compute metrics function
def compute_metrics(preds, targets):
    iou_threshold = 0.5
    total_iou, correct, total_pred, total_gt = 0, 0, 0, 0

    for pred, target in zip(preds, targets):
        pred_boxes = pred["boxes"]
        target_boxes = target["boxes"]

        if len(pred_boxes) > 0 and len(target_boxes) > 0:
            ious = box_iou(pred_boxes, target_boxes)
            max_ious, _ = ious.max(dim=1)
            total_iou += max_ious.sum().item()
            correct += (max_ious > iou_threshold).sum().item()

        total_pred += len(pred_boxes)
        total_gt += len(target_boxes)

    precision = correct / total_pred if total_pred > 0 else 0
    recall = correct / total_gt if total_gt > 0 else 0
    mean_iou = total_iou / correct if correct > 0 else 0
    return precision, recall, mean_iou

for epoch in range(num_epochs):
    model.train()
    total_loss, total_classifier_loss, total_box_loss = 0, 0, 0

    for images, targets in train_loader:
        images = [img.to(device) for img in images]
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        optimizer.zero_grad()
        loss_dict = model(images, targets)
        loss = sum(loss for loss in loss_dict.values())

        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        total_classifier_loss += loss_dict['loss_classifier'].item()
        total_box_loss += loss_dict['loss_box_reg'].item()

    scheduler.step()

    # Evaluation mode for metrics calculation
    model.eval()
    precision, recall, mean_iou = 0, 0, 0
    with torch.no_grad():
        for images, targets in valid_loader:
            images = [img.to(device) for img in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
            outputs = model(images)

            pred_list = [
                {"boxes": output["boxes"].detach().cpu(), "scores": output["scores"].detach().cpu()}
                for output in outputs
            ]
            target_list = [
                {"boxes": target["boxes"].detach().cpu()}
                for target in targets
            ]

            p, r, iou = compute_metrics(pred_list, target_list)
            precision += p
            recall += r
            mean_iou += iou

    precision /= len(valid_loader)
    recall /= len(valid_loader)
    mean_iou /= len(valid_loader)

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}, Class Loss: {total_classifier_loss:.4f}, Box Loss: {total_box_loss:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, IoU: {mean_iou:.4f}")

    # Print potential issues with training
    if total_loss > 2.0:
        print("⚠️ High loss detected. Consider reducing learning rate or adding data augmentation.")
    if precision < 0.5:
        print("⚠️ Low precision. The model might struggle with false positives.")
    if recall < 0.5:
        print("⚠️ Low recall. The model might be missing objects.")
    if mean_iou < 0.5:
        print("⚠️ Low IoU. Consider improving anchor box selection or increasing dataset size.")

print("✅ Training complete!")
```

```python
def collate_fn(batch):
    """Collate function for variable-size bounding boxes in batches."""
    images, targets = zip(*batch)  # Unzip batch into images and targets

    # Pad the 'boxes' tensor to the maximum size in the batch
    max_num_boxes = max(len(t["boxes"]) for t in targets)
    padded_boxes = [
        torch.cat([t["boxes"], torch.zeros(max_num_boxes - len(t["boxes"]), 4)], dim=0)
        if len(t["boxes"]) < max_num_boxes else t["boxes"]
        for t in targets
    ]

    # Pad the 'labels' tensor to the maximum size in the batch
    padded_labels = [
        torch.cat([t["labels"], torch.zeros(max_num_boxes - len(t["labels"]), dtype=torch.int64)], dim=0)
        if len(t["labels"]) < max_num_boxes else t["labels"]
        for t in targets
    ]

    # Update targets with padded tensors
    targets = [{"boxes": b, "labels": l} for b, l in zip(padded_boxes, padded_labels)]

    return list(images), list(targets)  # Convert to lists for the model

# Update the training loop to use this collate_fn
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=collate_fn)
```

```python
import os
import torch
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import xml.etree.ElementTree as ET

class VOCDataset(Dataset):
    def __init__(self, image_dir, annotation_dir, transform=None):
        self.image_dir = image_dir
        self.annotation_dir = annotation_dir
        self.image_filenames = [
            f for f in os.listdir(image_dir) if f.endswith(".jpg")
            and os.path.exists(os.path.join(annotation_dir, f.replace(".jpg", ".xml")))
        ]  # ✅ Only keep images that have annotations

        self.transform = transform

        # ✅ Predefined Label Mapping for consistency
        self.label_map = {
            "mobile_phone": 1, "walking_stick": 2, "switch_board": 3, "keyhole": 4,
            "key": 5, "door": 6, "window": 7, "stairs": 8, "crosswalk": 9,
            "traffic_light": 10, "road": 11, "house": 12, "building": 13, "car": 14,
            "bus": 15, "motorbike": 16, "cycle": 17
        }
        self.inverse_label_map = {v: k for k, v in self.label_map.items()}  # ✅ Reverse mapping

    def parse_voc_xml(self, xml_path):
        """Parses a VOC XML file and extracts bounding boxes & labels."""
        tree = ET.parse(xml_path)
        root = tree.getroot()

        boxes = []
        labels = []

        for obj in root.findall("object"):
            label = obj.find("name").text
            if label not in self.label_map:  # ✅ Ignore unknown labels
                continue

            bbox = obj.find("bndbox")
            xmin = float(bbox.find("xmin").text)
            ymin = float(bbox.find("ymin").text)
            xmax = float(bbox.find("xmax").text)
            ymax = float(bbox.find("ymax").text)

            boxes.append([xmin, ymin, xmax, ymax])
            labels.append(self.label_map[label])  # ✅ Use predefined label index

        # ✅ Return empty tensors if no objects are found
        if not boxes:
            return torch.empty((0, 4), dtype=torch.float32), torch.empty((0,), dtype=torch.int64)

        return torch.tensor(boxes, dtype=torch.float32), torch.tensor(labels, dtype=torch.int64)

    def __getitem__(self, idx):
        img_filename = self.image_filenames[idx]
        img_path = os.path.join(self.image_dir, img_filename)
        xml_path = os.path.join(self.annotation_dir, img_filename.replace(".jpg", ".xml"))

        # ✅ Load image safely
        image = Image.open(img_path).convert("RGB")

        # ✅ Parse annotation safely
        boxes, labels = self.parse_voc_xml(xml_path)

        target = {"boxes": boxes, "labels": labels}

        if self.transform:
            transformed = self.transform(image, target)
            image = transformed["image"]
            target = {"boxes": transformed["boxes"], "labels": transformed["labels"]}

        return image, target

    def __len__(self):
        return len(self.image_filenames)

def collate_fn(batch):
    """Collate function for variable-size bounding boxes in batches."""
    images, targets = zip(*batch)
    return images, targets
```
